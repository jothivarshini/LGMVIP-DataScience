# **Importing Modules**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
from google.colab.patches import cv2_imshow
```

# **Reading and displaying original image**

```python
img=cv2.imread('cam.jpg')
cv2_imshow(img)
```

# **Converting to Grayscale**

```python
gray_img=cv2.imread("cam.jpg",0)
cv2_imshow(gray_img)
```

# **Inverting Image**

```python
inverted_img = 255 - gray_img
cv2_imshow(inverted_img)
```

# **Blurring the Image** 

```python
 blur_img = cv2.GaussianBlur(inverted_img,(15,15),0)
 cv2_imshow(blur_img)
```

# **Presenting the Final Pencil Sketch**

```python
inverted_blur_img = 255 - blur_img
pencil_sketch = cv2.divide(gray_img,inverted_blur_img,scale = 256.0)
cv2_imshow(pencil_sketch)
```
